```
#Problem 2
#construct the fibonacci sequence - generated by adding the previous two terms 
#range(4000000)
#find sum of the even-valued terms 

def fibon(limit):
  a = 0 
  b = 1
  while a < limit: 
    yield a
    a = b 
    b = a + b 
print(sum(i for i in fibon(4000000) if i % 2 == 0))


```

```
#Problem 3
#The prime factors of 13195 are 5, 7, 13 and 29.
#What is the largest prime factor of the number 600851475143 ?

#for values that are only divisible by 1 and itself 
#Max prime factor
#largest number 600851475143

n = 600851475143
i = 2
while i*i < n:
  while n%i == 0:
    n = n/i 
  i = i + 1 

print(n)
```

```
#Problem 4
#A palindromic number reads the same both ways. The largest palindrome made from 
#the product of two 2-digit numbers is 9009 = 91 × 99.

#Find the largest palindrome made from the product of two 3-digit numbers.

#create a function for palindromic 
#Want to use two 3-digits (100,999)
#answer bust be read the same both ways 

palindromes = []

for x in range(100,999):
  for i in range(100,999):
    num = str(i * x).split()[0]
    if len(num) == 6:
      if num[0] == num[-1] and num[1] == num[-2] and num[2] == num[-3]:
        palindromes.append(i * x) 

print(max(palindromes))
```

```
#Problem 5
#2520 is the smallest number that can be divided by each of the numbers from 1 
#to 10 without any remainder.

#What is the smallest positive number that is evenly divisible by all of the 
#numbers from 1 to 20?

#for loop to find smallest postive number 
#must be evenly divisible by numbers in range(1,21)
#smallest number 

#start at 2520 because up to this point it is divisible by numbers 1-10 
#thus, start at 11

#function to find divisibility
def check(n):               #create function 
  for i range(11,21):       #for every value in range 11-20
    if n % i == 0:            #if it's evenly divisible 
      continue                  #then continue 
    else:                     #Otherwise 
      return False              #return false
  return True               #if it passes all above pass true 


#main function 
x = 2520                    #starting value 

while not check(x):         #False values are checked through loop by being added 2520 
  x += 2520                 #adds a number to x 

print(x)
```

```
#Problem 6

#The sum of the squares of the first ten natural numbers is,
#1^2+2^2+...+10^2=385 

#The square of the sum of the first ten natural numbers is,
#(1+2+...+10)^2=55^2=3025 

#Hence the difference between the sum of the squares of 
#the first ten natural numbers and the square of the sum is 3025−385=2640.

#Find the difference between the sum of the squares of the first one hundred 
#natural numbers and the square of the sum.

#Def and Find the sum of the squares of the first 100 natural numbers 
#Then, def and find the suqare of the sum of the first 100 natural numbers
#And find the difference of the numbers 

def sum_of_sqs(limit):              #def function with limit(100)
  total = 0                         #
  for i in range(limit+1):          #for every value in the range including 100
    total += i**2                   #total adds the next value squared 
  return total                      #when for loop is done it will return total  

def square_of_sums(limit):          #def func with limit(100)
  total = 0             
  for i in range(limit+1):          #for every value in the range including 100
    total += i                      #the value is added to the total 
  return total**2                   #sum of the total is squared 

s1 = sum_of_sqs(100)
s2 = square_of_sums(100)

print("Answer:" + str(s2-s1))       #difference 


```

```
#Problem 9

#A Pythagorean triplet is a set of three natural numbers, a < b < c, for which,
#a2 + b2 = c2

#For example, 3^2 + 4^2 = 9 + 16 = 25 = 52.

#There exists exactly one Pythagorean triplet for which a + b + c = 1000.
#Find the product abc.

def prod_abc(target):
  for a in range(1, target):
    for b in range(a, target):
      ab_sq = a * a + b * b 
      approx_c = target -a -b 
      if approx_c <= b:
        continue
      if approx_c * approx_c == ab_sq:
        return (a, b, approx_c)
  assert False

target = 1000

a, b, c = prod_abc(target)
print(a, b, c, a*a, b*b, c*c)
```
